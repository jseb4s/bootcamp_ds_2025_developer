```python
import pandas as pd
import requests

url = "https://www.datos.gov.co/resource/sv9t-4pvf.json"
response = requests.get(url)

# JSON a DataFrame
df = pd.DataFrame(response.json())
df.head()
```

```python
df_filtrado = df.loc[(df["municipio"] == "CALI")]
df_filtrado.head()
```

```python
def limpiar_columnas(df, columnas):
    for col in columnas:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(r"[\$, ]", "", regex=True)
            .str.replace(",", "")
            .astype(float)
        )
    return df
```

```python
df_filtrado = limpiar_columnas(df_filtrado, ["presupuesto_inicial_millones", "presupuesto_definitivo", "total_compromisos_millones",
                                             "total_obligaciones_millones", "pagos_efectivos_a_fecha_de"])
df_filtrado.head()
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)  # Muestra 2 decimales
df_filtrado.describe()
```

```python
df_filtrado["compromisos_presupuesto"] = df_filtrado["compromisos_presupuesto"].astype(float)
df_filtrado["pagos_presupuesto_definitivo"] = df_filtrado["pagos_presupuesto_definitivo"].astype(float)
df_filtrado["pagos_compromisos"] = df_filtrado["pagos_compromisos"].astype(float)
```
